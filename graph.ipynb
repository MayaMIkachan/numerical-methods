```python
import json
import matplotlib.pyplot as plt

stats = None
with open('report.json') as f:
    stats = json.load(f)

montecarlo_times = [statistic['MonteCarlo']['ExecutionTime']
                    for statistic in stats]
montecarlo_invocations = [statistic['MonteCarlo']
                          ['InvokeCount'] for statistic in stats]
quadrature_times = [statistic['Quadrature']['ExecutionTime']
                    for statistic in stats]
quadrature_invocations = [statistic['Quadrature']
                          ['InvokeCount'] for statistic in stats]

n = [i+1 for i in range(len(montecarlo_times))]
```

```python
plt.plot(n, montecarlo_invocations, label='MonteCarlo')
plt.plot(n, quadrature_invocations, label='Quadrature')
plt.title('Зависимость количества вызовов функции\n от кратности интеграла')
plt.legend()
```

```python
plt.plot(n, montecarlo_times, label='MonteCarlo')
plt.plot(n, quadrature_times, label='Quadrature')
plt.title('Зависимость времени выполнения\n от кратности интеграла')
plt.legend()
```
